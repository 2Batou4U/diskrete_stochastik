## Übungsblatt 3

#### Aufgabe 1a

Zum Berechnen des Erwartungswerts einer Poisson-Verteilten Zufallsvariable $ X $ müssen wir zunächst wissen, was die Formel für diesen ist. Dieser ist nach Beispiel 5.7 des Skripts:

$$
\mathbb{E}X = \sum_{k=0}^{\infty} k \frac{\lambda^{k}}{k!} e^{-\lambda}
$$

Da $ e^{-\lambda} $ von k unabhängig ist, können wir es zur besseren Übersicht aus der Summe herausholen. 

Außerdem können wir das $ k $ wegkürzen, indem wir aus dem Nenner $ k! $ ein $ (k-1)! $ machen.

$$
\mathbb{E}X = e^{-\lambda} \sum_{k=0}^{\infty} \frac{\lambda^{k}}{(k-1)!} 
$$

Jetzt holen wir uns noch eins der $ \lambda^{k} $ aus der Summenformel raus und machen daraus dann zum einen $ \lambda $ sowie
$ \lambda^{k-1} $. Da wir nun bei $ k-1 $ nicht mehr bei $ k=0 $ anfangen können, setzen wir die Summe auf $ 1 $. Hieraus folgt dann:

$$
\mathbb{E}X = \lambda e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^{k - 1}}{(k-1)!} 
$$

Die Summenformel ist nun die Darstellung einer Taylorreihe und kann umgeformt werden:

$$
\mathbb{E}X = \lambda e^{-\lambda} e^{\lambda} 
$$

Die Eulerschen Zahlen kürzen sich weg und somit erreichen wir die endgültige Formel für den Erwartungswert von $ X $:

$$
\mathbb{E}X = \lambda
$$

**Somit ist der Erwartungswert einer Zufallsvariablen $ X $ gleich $ \lambda $.**

---

Nun berechnen wir den Erwartungswert von $ \mathbb{E}X(X-1) $.

Hierzu holen wir uns die gleiche Formel heran wie bereits bei $ X $ und setzen die entsprechenden Werte ein:

$$
\mathbb{E}X(X-1) = \sum_{k=1}^{\infty} k(k-1) \frac{\lambda^{k}}{k!} e^{-\lambda}
$$

Auch hier können wir uns das $ e^{-\lambda} $ aus der Summe holen, das $ k(k-1) $ ist quasi $ k! / (k-2)! $ und wir können es kürzen, indem $ k! $ zu $ (k-2)! $ wird. Da wir nun $ k-2 $ für die Formel brauchen, fängt die Summe bei $ k = 2 $ an:

$$
\mathbb{E}X(X-1) = e^{-\lambda} \sum_{k=2}^{\infty} \frac{\lambda^{k}}{(k-2)!} 
$$

Jetzt holen wir uns ein $ \lambda^{2} $ raus, um die Taylorreihe wieder anwenden zu können:

$$
\mathbb{E}X(X-1) = \lambda^{2} e^{-\lambda} \sum_{k=2}^{\infty} \frac{\lambda^{k-2}}{(k-2)!} 
$$

Mir fällt auf, dass ich vorhin nicht richtig erklärt haben, wie:

$$
\sum_{k=2}^{\infty} \frac{\lambda^{k-2}}{(k-2)!} 
$$

zu

$$
e^{\lambda}
$$

wird.

Die Formel für die Taylorreihe:

$$
e^{a} = \sum_{n=0}^{\infty} \frac{a^{n}}{n!} 
$$

Wir können $ (k-2) $ als unser $ a $ sehen und somit die Gleichung anwenden.

Anyway, aus der Formel wird dann:

$$
\mathbb{E}X(X-1) = \lambda^{2} e^{-\lambda} e^{\lambda}
$$

das kürzt sich wieder weg zu:

$$
\mathbb{E}X(X-1) = \lambda^{2}
$$

**Der Erwartungswert von $ \mathbb{E}X(X-1) $ ist also gleich $ \lambda^{2} $.**

#### Aufgabe 1b

Okay, die Formel für den Erwartungswert ist:

$$
\mathbb{V}X = \mathbb{E}(X - \mathbb{E}X)^2
$$

Der Einfachheitshalber können wir dies wie folgt schreiben:

$$
\mathbb{V}X = \mathbb{E}X^2 - (\mathbb{E}X)^2
$$

(Quelle Beispiel 5.24)

Zuletzt hatten wir in Aufgabe 1a den Erwartungswert von $ \mathbb{E}X $, sowie von $ \mathbb{E}X(X-1) $ berechnet.

Wir wissen, dass wir auf $ \mathbb{E}X^2 $ kommen müssen. Hierfür können wir $ \mathbb{E}X(X-1) + \mathbb{E}X$ berechnen, da diese equivalent zu $ \mathbb{E}X^2 $ sind. Da wir das Ergebnis von den beiden kennen ($ \lambda^{2} $, sowie $ \lambda $), wissen wir, dass $ \mathbb{E}X^2 $ den Erwartungswert $ \lambda^{2} + \lambda $ haben muss. Der Erwartungswert von $ X $ kommt noch dazu und wir verenden mit der Formel für die Varianz einer Poisson-Verteilten Zufallsvariable:

$$
\mathbb{V}X = \lambda^{2} + \lambda - \lambda^{2}
$$

was sich wiederum zu folgendem zusammenfassen lässt:

$$
\mathbb{V}X = \lambda
$$

**Somit ist die Varianz von $ X $ ebenfalls gleich $ \lambda $.**

#### Aufgabe 1c

Um zu zeigen, dass die Summe zweier unabhängiger Poisson-verteilter Zufallsvariablen wieder Poisson-verteilt ist, kann man die Eigenschaften der Poisson-Verteilung und die Unabhängigkeit der Variablen nutzen.

Seien $ X $ und $ Y $ zwei unabhängige Poisson-Verteilte Zufallsvariablen mit den Parametern $ \mu > 0 $ und $ \lambda > 0 $. Das bedeutet, dass für $X$ und $Y$ gilt:

$$ 
P(X = k) = \frac{e^{-\mu} \mu^k}{k!} \quad \text{und} \quad P(Y = l) = \frac{e^{-\lambda} \lambda^l}{l!} 
$$

für $k, l = 0, 1, 2, \ldots$.

Wir interessieren uns für die Verteilung der Summe $Z = X + Y$. Um die Wahrscheinlichkeit $P(Z = m)$ zu bestimmen, nutzen wir die Tatsache, dass $X$ und $Y$ unabhängig sind. Daher ist die Wahrscheinlichkeit, dass $Z = m$, gleich der Summe der Wahrscheinlichkeiten, dass $X = k$ und $Y = m - k$:

$$ 
P(Z = m) = \sum_{k=0}^{m} P(X = k) \cdot P(Y = m - k) 
$$

Setzen wir die Poisson-Verteilungen für $X$ und $Y$ ein, erhalten wir:

$$ 
P(Z = m) = \sum_{k=0}^{n} \left( \frac{e^{-\mu} \mu^k}{k!} \right) \left( \frac{e^{-\lambda} \lambda^{m-k}}{(m-k)!} \right) 
$$

Dies lässt sich umformen zu:

$$ 
= e^{-(\mu + \lambda)} \sum_{k=0}^{m} \frac{\mu^k \lambda^{m-k}}{k!(n-k)!} 
$$

Man sieht, dass dies der binomische Lehrsatz angewendet auf $(\mu + \lambda)^n$ ist. Daher können wir vereinfachen zu:

$$ P(Z = m) = e^{-(\mu + \lambda)} \frac{(\mu + \lambda)^m}{m!} $$

**Also ist $Z$ Poisson-verteilt mit dem Parameter $\mu + \lambda$.**

#### Aufgabe 3a

Da $ X_n $ die Werte $ -n, \ldots, -1, 0, 1, \ldots, n $ jeweils mit gleicher Wahrscheinlichkeit annimmt, ist die Wahrscheinlichkeit für jeden Wert $ \frac{1}{2n+1} $, weil es insgesamt $ 2n+1 $ mögliche Werte gibt (von $-n$ bis $n$ einschließlich $0$).

Der Erwartungswert $ \mathbb{E}X_n $ ist definiert als die Summe der Produkte jedes Werts mit seiner Wahrscheinlichkeit:

$$ \mathbb{E}X_n] = \sum_{k=-n}^{n} k \cdot P(X_n = k) $$

**Da die Werte symmetrisch um $0$ verteilt sind, heben sich die positiven und negativen Werte gegenseitig auf. Daher ist der Erwartungswert $ \mathbb{E}X_n = 0 $.**

---

Die Varianz $ V(X_n) $ ist definiert als der Erwartungswert der quadrierten Abweichungen vom Erwartungswert. Da $ \mathbb{E}X_n = 0 $, vereinfacht sich dies zu:

$$ V(X_n) = \mathbb{E}X_n^2 = \sum_{k=-n}^{n} k^2 \cdot P(X_n = k) $$

Wiederum ist $ P(X_n = k) = \frac{1}{2n+1} $ für alle $ k $.

$$ V(X_n) = \frac{1}{2n+1} \sum_{k=-n}^{n} k^2 $$

Da $ k^2 $ für $ k $ und $ -k $ gleich ist, können wir die Formel vereinfachen:

$$ V(X_n) = \frac{2}{2n+1} \sum_{k=1}^{n} k^2 $$

Die Summe der Quadrate der ersten $ n $ natürlichen Zahlen ist, diese Formel ist aus dem Internet geklaut, aber sie scheint zu stimmen:

$$ \sum_{k=1}^{n} k^2 = \frac{n(n+1)(2n+1)}{6} $$

Setzen wir dies in die Formel für $ V(X_n) $ ein:

$$ V(X_n) = \frac{2}{2n+1} \cdot \frac{n(n+1)(2n+1)}{6} $$

Vereinfachen wir dies:

$$ V(X_n) = \frac{n(n+1)}{3} $$

**So ist der Erwartungswert $ \mathbb{E}X_n = 0 $ und die Varianz $ V(X_n) = \frac{n(n+1)}{3} $.**

#### Aufgabe 3b

Um $ P(|X_n| \geq \frac{n}{2}) $ zu berechnen, müssen wir die Wahrscheinlichkeit finden, dass der Betrag der Zufallsvariablen $ X_n $ gleich oder größer als $ \frac{n}{2} $ ist. 

Da $ X_n $ die Werte von $ -n $ bis $ n $ mit gleicher Wahrscheinlichkeit annimmt, betrachten wir alle Werte von $ X_n $, die außerhalb des Bereichs $ -\frac{n}{2} $ bis $ \frac{n}{2} $ liegen. 

1. **Bereich der relevanten Werte**: Die Werte von $ X_n $, die $ |X_n| \geq \frac{n}{2} $ erfüllen, sind $ -n, -n+1, \ldots, -\frac{n}{2} - 1 $ und $ \frac{n}{2} + 1, \frac{n}{2} + 2, \ldots, n $. 

2. **Anzahl der relevanten Werte**: Es gibt $ n - \frac{n}{2} = \frac{n}{2} $ Werte in jedem Bereich (negativ und positiv), also insgesamt $ 2 \times \frac{n}{2} = n $ Werte.

3. **Gesamtanzahl der möglichen Werte**: Es gibt insgesamt $ 2n + 1 $ mögliche Werte für $ X_n $.

4. **Berechnung der Wahrscheinlichkeit**: Da jeder Wert mit gleicher Wahrscheinlichkeit auftritt, ist die Wahrscheinlichkeit für jeden einzelnen Wert $ \frac{1}{2n + 1} $. 

   Die Wahrscheinlichkeit, dass $ |X_n| \geq \frac{n}{2} $, ist daher $ \frac{n}{2n + 1} $.

**Zusammengefasst ist $ P(|X_n| \geq \frac{n}{2}) = \frac{n}{2n + 1} $.**

#### Aufgabe 3c

In Aufgabe 3b wurde berechnet $ P(|X_{100}| \geq \frac{100}{2}) $. Für $ n = 100 $, wird diese Wahrscheinlichkeit:

$$ P(|X_{100}| \geq 50) = \frac{100}{2 \times 100 + 1} = \frac{100}{201} $$

#### Tschebyschev-Ungleichung

Die Tschebyschev-Ungleichung lautet:

$$ P(|X - \mu| \geq k) \leq \frac{\sigma^2}{k^2} $$

Für unsere Zufallsvariable $ X_{100} $ mit $ \mu = E[X_{100}] = 0 $ (da symmetrisch verteilt) und $ \sigma^2 = V(X_{100}) $. Aus Aufgabe 3a wissen wir, dass $ V(X_{100}) = \frac{100 \times 101}{3} $.

Setzen wir $ k = 50 $ in die Tschebyschev-Ungleichung ein:

$$ P(|X_{100}| \geq 50) \leq \frac{\frac{100 \times 101}{3}}{50^2} $$

Berechnen wir den rechten Teil der Ungleichung:

$$ \frac{\frac{100 \times 101}{3}}{50^2} = \frac{100 \times 101}{150^2} = 0.449 $$

Jetzt vergleichen wir die beiden Ergebnisse:

- Direkte Berechnung: $ P(|X_{100}| \geq 50) = \frac{100}{201} $
- Tschebyschev-Abschätzung: $ P(|X_{100}| \geq 50) \leq 0.449 $
